```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gmplot
```

Read CSV file

```python
df=pd.read_csv('Dataset_1_Datafiniti_Hotel_Reviews_Jun19.csv')
df
```

1. Top 5 hotels based on the review ratings in each state.

```python
def one():
    print(df.groupby(['province']).apply(lambda x: x.head().sort_values(['reviews.rating'], ascending=False)))
```

2. Display the list of hotels sorted on the number of reviews.

```python
def two():
    print(df['address'].value_counts())
```

3. Display the reviews for any hotel prompted by the User. In case no review is found, an appropriate message should be displayed.

```python
def three():
    user = input("Enter address of the hotel:- ")
    a = df[['reviews.username','reviews.title','reviews.text','reviews.rating']].loc[df['address'] == user]
    if(a.empty):
        print('No reviews found')
    print(a)
```

4. Display the list of hotels in each category.

```python
def four():
    print(df.groupby(['categories']).apply(lambda x: x.sort_values(['reviews.rating'], ascending=False)))
```

5. Display a count of hotels for each state.

```python
def five():
    print(df.groupby('province').address.nunique())
```

6. Search for any hotel and display all the relevant information.

```python
def six():
    user = input("Enter address of the hotel:- ")
    a = df.loc[df['address'] == user]
    if(a.empty):
        print('No reviews found')
    print(a)
```

7. Top five cities whose residents wrote the maximum reviews.

```python
def seven():
    print(df['reviews.userCity'].value_counts().head())
```

8. Display list of hotels having rating greater than 4 by atleast 3 users.

```python
def eight():
    newdf = df[df['reviews.rating'] > 4]
    df1 = newdf.groupby('address').count()
    fdf = df1[df1['reviews.rating'] >= 3]
    print(fdf['reviews.rating'])
```

9. Display list of all hotels having more than 4 reviews

```python
def nine():
    df1 = df.groupby('address').count()
    df2 = df1[df1['reviews.username'] > 4]
    print(df2['reviews.username'])
```

10. Display relevant information about hotels in a city asked by the user

```python
def ten():
    user = input("Enter city:- ")
    a = df.loc[df['city'] == user]
    if(a.empty):
        print('No reviews found')
    print(a.groupby('address').apply(lambda x: x.sort_values(['reviews.rating'], ascending=False)))
```

11. Display list of usernames for all hotel with there review rating

```python
def eleven():
    a=df.groupby(['address','reviews.username','reviews.rating'])
    print(a.sourceURLs.unique())
```

12. To count top 10 websites used

15. Display the ratings of a particular hotel in bar graph

```python
def fifteen():
    user = input("Enter address of the hotel:- ")
    a = df.loc[df['address'] == user]
    if(a.empty):
        print('No reviews found')
    c = a['reviews.rating'].value_counts(sort=False).tolist()
    b1 = a['reviews.rating'].unique().tolist()
    b2 = sorted(b1)
    print(plt.bar(b2,c))
    print("Graph will be plotted at the end of queries")
```
